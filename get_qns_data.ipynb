#Pulling Queens Complaints from Full 311 Complaint Set

- The "Borough" listing in the 311 complaint set is unreliable - there are over 1 million "Unspecified" listings making up ~ 10% of the data
- The solution I've chosen, while imperfect, is to pull Queens data based on zipcode.

###What this notebook does
1. Load the full 311 data
2. Clean up the zip codes (there are some with the 4 digit suffix, others that aren't full zip codes, and some combinations of strings and numbers ex: "NJ 03").
3. Select zip codes that are from Queens only

The issues with this method are that it assumes the zip codes are the correct zip codes for the complaint.  Additionally, the zip codes can span borough boundaries, particularly between Queens and Brooklyn.

### Pull out Queens via zipcodes; load in a list of zip codes in Queens

```python
zip = list(pd.read_csv('queens_zipcodes.csv'))
```

```python
df[df['Incident Zip'].notnull()]['Incident Zip'].astype(str)
```

```python
# Clean up zips a bit, remove any digits greater than 5, remove any zips that aren't numbers
s = df['Incident Zip']
df.zip = s.map(lambda x: str(x)[:5])
zip2 = [np.nan if len(re.sub('\D','',str(x)))!=5 else x for x in df.zip]
df.zip2 = zip2
df['zip2'] = pd.Series(zip2, index=df.index)
df['zip2'].astype(float)
```

```python
#Pull out rows with zipcodes in Queens zipcode list and set "borough" column to queens
qns_ind = [True if x in zip else False for x in df['zip2']]
qns_data = df[qns_ind]
qns_data.loc[:,'Borough']='queens'
qns_data.reset_index()
```
