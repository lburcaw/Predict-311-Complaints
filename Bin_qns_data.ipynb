```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('qns_complaints.csv')
```

### Pull out main 4 complaint types:  Building, Street, Vermin, Noise

```python
complts = df['Complaint Type'].value_counts()
```

```python
#Get an idea of the distribution of complaints
prop = complts[:100]/sum(complts[:100])
prop.plot()
```

```python
tot = np.cumsum(prop)
tot.plot()
```

```python
df['Complaint Type']
```

```python
# Define complaint classes
bldg_cplt = ['DOOR/WINDOW','ELECTRIC','FLOORING/STAIRS','HEAT/HOT WATER','HEATING','OUTSIDE BUILDING','PAINT/PLASTER',
             'PLUMBING']
stswk_cplt = ['Illegal Parking','Blocked Driveway','Sidewalk Condition','Snow','Street Condition','Street Light Condition',
              'Sweeping/Inadequate','Sweeping/Missed','Sweeping/Missed-Inadequate']
vm_cplt = ['Rodent']
vm_desc = ['VERMIN','PESTS']
nz_cplt = ['Noise','Noise - Commercial','Noise - Helicopter','Noise - House of Worship','Noise - Park',
           'Noise - Street/Sidewalk','Noise - Vehicle']
```

```python
# Pull out rows corresponding to particular complaint type
bldg_ind = [True if x in bldg_cplt else False for x in df['Complaint Type']]
stswk_ind = [True if x in stswk_cplt else False for x in df['Complaint Type']]
vm_ind1 = [True if x in vm_cplt else False for x in df['Complaint Type']]
vm_ind2 = [True if x in vm_desc else False for x in df['Descriptor']]
nz_ind = [True if x in nz_cplt else False for x in df['Complaint Type']]
```

```python
vm_ind = [any(tup) for tup in zip(vm_ind1,vm_ind2)]
```

```python
# Get an idea of how unbalanced the labels are
counts = [sum(bldg_ind),sum(stswk_ind),sum(vm_ind),sum(nz_ind)]
[float(x)/sum(counts) for x in counts]
```

```python
#Define complaint dataframes
df_blg_only = df[bldg_ind]
df_stswk_only = df[stswk_ind]
df_vm_only = df[vm_ind]
df_nz_only = df[nz_ind]
```

```python
#top 5 names
bldg_vcounts_split = df_blg_only['Complaint Type'].value_counts()
stswk_vcounts = df_stswk_only['Complaint Type'].value_counts()
vm_vcounts = df_vm_only['Complaint Type'].value_counts()
nz_vcounts = df_nz_only['Complaint Type'].value_counts()
```

```python
top5_bldg_ind = [True if x in bldg_vcounts_split.index.values[:7] else False for x in df['Complaint Type']]
top5_stswk_ind = [True if x in stswk_vcounts.index.values[:5] else False for x in df['Complaint Type']]
top5_nz_ind = [True if x in nz_vcounts.index.values[:5] else False for x in df['Complaint Type']]
```

```python
df_bldg_binned = df[top5_bldg_ind]
df_stswk_binned = df[top5_stswk_ind]
df_nz_binned = df[top5_nz_ind]
```

```python
# Save to csv
df_vm_only.to_csv('vm_complaints.csv')
df_bldg_binned.to_csv('bldg_complaints.csv')
df_stswk_binned.to_csv('stswk_complaints.csv')
df_nz_binned.to_csv('nz_complaints.csv')
```

```python
# See how unbalanced labels are
top5_counts = [sum(top5_bldg_ind),sum(top5_stswk_ind),sum(vm_ind),sum(top5_nz_ind)]
prop_top5_counts = [float(x)/sum(top5_counts) for x in top5_counts]
```

```python
# Label top 5 complaints in each type (not used in final product)
bldg_val_temp = [bldg_vcounts_split.values[0]+bldg_vcounts_split.values[3],bldg_vcounts_split.values[1],
                 bldg_vcounts_split.values[2],bldg_vcounts_split.values[4],sum(bldg_vcounts_split.values[5:7])]
```

```python
bldg_vcounts = pd.Series(bldg_val_temp,index=['Heating','Plumbing','Electric','Paint/Plaster','Structural'])
bldg_vcounts
```

```python
prop_bldg_vcounts = pd.Series([float(x)/sum(top5_bldg_ind) for x in bldg_vcounts.values[:5]])
prop_stswk_vcounts = pd.Series([float(x)/sum(top5_stswk_ind) for x in stswk_vcounts.values[:5]])
prop_nz_vcounts = pd.Series([float(x)/sum(top5_nz_ind) for x in nz_vcounts.values[:5]]) 
prop_vm_vcounts = pd.Series([float(x)/sum(vm_ind) for x in vm_vcounts.values[:5]])
```
